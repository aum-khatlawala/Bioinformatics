```python
start = 0
end = 0
with open('FJ959345.txt', 'r') as file:
    i = 0
    for line in file.readlines():
        if(line[0:8] == "FEATURES"):
            start = i + 1
        if(line[0:6] == "ORIGIN"):
            end = i
        i = i + 1
with open('FJ959345.txt', 'r') as file:
    ans = "".join(list(file.readlines())[start:end]).replace('\n', ' ')
    ans = ans.replace(' ', '-')
    ans = ans.replace("Location/Qualifiers", "")
arr = ans.split('/')
# arr = ans.split()
arr=[i for i in arr if i!='']
# for i in arr:
#     if (i.count('"') == 1):
#         for x in arr+i:
#             print(x)
def split(seq):
    for item in seq:
        a, b, c = item.partition('"')
        d, e, f = c.partition('"')
        yield a+b+d+e
        yield f
answer = list(split(arr))
answer=[i for i in answer if i!='']
answer=[i for i in answer if i!='----------------------']
# print(answer)
# for i in answer:
#     i.replace('-', ' ')
# print(replace(string, char))


for i in answer:
    i.replace("----------------------", "")
#     print(i)
final_list = []
i = 0
while i < len(answer) - 1:
    if (answer[i][0] == '-'):
        buffer = answer[i] + ": "
        j = i + 1
        while answer[j][0] != '-':
            buffer = buffer + answer[j]
            j += 1
        final_list.append(buffer)
        i = j
```

```python
for i in range(len(final_list)):
    final_list[i] = final_list[i].replace("----------------------", "")
    final_list[i] = final_list[i].replace("----------", " ")
    final_list[i] = final_list[i].replace("------", " ")
    final_list[i] = final_list[i].replace("-", " ")
    
for i in range(len(final_list)):
    print("Table entry:", i + 1)
    print(final_list[i])
    print("\n")
```
