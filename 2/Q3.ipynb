### 3. It is assumed that A = T and G = C in a genome. Compute G+C content, (G+C)/(A+T+G+C), deviation from A = T, (A-T)/(A+T) and deviation from G = C, (G-C)/(G+C) to confirm. Use a non-overlapping sliding window of size 50Kb.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
with open('e-coli.txt', 'r') as file:
    whole_genome_string = file.read().replace('\n', '')
print(len(whole_genome_string))
n = math.ceil(len(whole_genome_string) / 50000)
a = np.zeros(n)
b = np.zeros(n)
c = np.zeros(n)
sets_of_50k = [[] for x in range(n)]
for i in range(len(whole_genome_string)):
    index = math.floor(i / 50000)
    sets_of_50k[index].append(whole_genome_string[i])

i = 0
for sample in sets_of_50k:
    G = sample.count('G')
    C = sample.count('C')
    A = sample.count('A')
    T = sample.count('T')
    a[i] = (G - C) / (G + C)
    b[i] = (A - T) / (A + T)
    c[i] = (G + C) / (A + T + G + C)
    i = i + 1
```

```python
print("GC Skew:")
plt.plot(a)
plt.axhline(y = 0)
```

```python
print("AT Skew:")
plt.plot(b)
plt.axhline(y = 0)
```

```python
print("GC Content:")
plt.plot(c)
```
