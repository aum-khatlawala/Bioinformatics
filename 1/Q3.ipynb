```python
from itertools import combinations
reverse_pairs = {
    "a":"t",
    "t":"a",
    "g":"c",
    "c":"g"
}
def checkpalindrome(DNA):
    reverse = []
    for i in DNA:
        reverse.append(reverse_pairs[i])
    if "".join(reverse)==DNA[::-1]:
        return 1
    else:
        return 0
dna_sequence = str(input("Enter DNA sequence: ")).lower().replace(" ", "")
lengths = [4, 6, 8]
rr_sites = {}
for length in lengths:
    for start, end in combinations(range(len(dna_sequence)+1), r = 2):
        if len(dna_sequence[start:end])==length:
            if checkpalindrome(dna_sequence[start:end]) == 1:
                if not end-start in rr_sites:
                    rr_sites[length]=[]
                rr_sites[length].append(dna_sequence[start:end])
print("\nPossible recognition sites: ")
for length, item in rr_sites.items():
    print("Length =", length)
    for i in item:
        print(i)
```
